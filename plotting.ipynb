### Plotting more than one curve

```python
## more than one plot
## signature plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
import matplotlib.pyplot as plt
x=[i for i in range(11)]
y=[i**2 for i in x]
u=[i for i in range(5,15)]
v=[i**3 for i in u]
plt.plot(x,y,u,v)# alternatively plt.plot(x,y) followed by plt.plot(u,v)
plt.show()
```

### Differentiating between plots

```python
## format
## mt = '[marker][line][color]'
## for various formats see https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
## Note different order for the two formats
fmt1='+:b'
fmt2='go-'
import matplotlib.pyplot as plt
x=[i for i in range(11)]
y=[i**2 for i in x]
u=[i for i in range(5,15)]
v=[i**3 for i in u]
plt.plot(x,y,fmt1,u,v,fmt2)# alternatively plt.plot(x,y) followed by plt.plot(u,v)
plt.show()
```

### Axes Properties

```python
### Labeling the axes
fmt1='+:b'
fmt2='go-'
import matplotlib.pyplot as plt
x=[i for i in range(11)]
y=[i**2 for i in x]
u=[i for i in range(5,15)]
v=[i**3 for i in u]
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x,y,fmt1,u,v,fmt2)# alternatively plt.plot(x,y) followed by plt.plot(u,v)
plt.show()
```

```python
### axis limits and ticks
fmt1='+:b'
fmt2='go-'
import matplotlib.pyplot as plt
x=[i for i in range(11)]
y=[i**2 for i in x]
u=[i for i in range(5,15)]
v=[i**3 for i in u]
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xlim((0,15))
plt.ylim((0,3000))
plt.xticks([i for i in range(16)])
plt.yticks([i for i in range(0,4000,1000)])
plt.title("My first plot")
plt.plot(x,y,fmt1,label="plot1")
plt.plot(u,v,fmt2,label="plot2")# alternatively plt.plot(x,y) followed by plt.plot(u,v)
## 0=best,1=upper right,2=upper left,3=lower left,4=lower right
plt.legend(loc=4)

plt.show()
```
