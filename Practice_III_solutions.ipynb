```python
#Problem 1
class Point:
    def __init__(self,x=0,y=0):
        assert (type(x)==int or type(x)==float) and (type(y)==int or type(y)==float), 'Bad Input!'
        self.x = x
        self.y = y
    
    def __str__(self):
        return '(' + str(self.x) + ','+ str(self.y)+ ')'
    
class Rectangle: 
    def __init__(self,p,q):
        assert type(p)==Point and type(q)==Point and p.x<=q.x and p.y<=q.y, 'Bad Input!'
        self.p = p  # be careful that we are aliasing here! 
        self.q = q  # but since none of the methods in Rectangle modifies p or q, then this aliasing is safe
        
    def __str__(self):
        return '(p=' + str(self.p)+','+'q='+ str(self.q) + ')'
    
    def height(self):
        return self.q.y - self.p.y
    
    def width(self):
        return self.q.x - self.p.x
    
    def perimeter(self):
        return 2*(self.height()+self.width())
    
    def area(self):
        return self.height()*self.width()
    
    def contains(self,pt):
        assert type(pt)==Point, 'Bad Input!'
        return (self.p.x<=pt.x<=self.q.x) and (self.p.y<=pt.y<=self.q.y)
    

p1 = Point()
p2 = Point(3)
p3 = Point(2.1,5)

R2 = Rectangle(p1,p2)

print(p1,p2,p3)

#p4 = Point('hi',4)  #Assertion Error

#R = Rectangle(Point(3,4), Point(1,2))  #Assertion Error



R = Rectangle(Point(1,2),Point(3.2,4))
print(R)
print(R.width())
print(R.height())
print(R.area())
print(R.perimeter())
print(R.contains(Point(1.5,3)))
print(R.contains(Point(10.5,3)))

#print(R.contains(3)) #Assertion Error
```

```python
#Problem 3

def merge(L,R):
    C = [] # initialize C to the empty list
    m = len(L)
    n = len(R)
    i = 0 # counter to traverse L
    j = 0 # counter to traverse R
    while i!=m and j!=n:
        if L[i] < R[j]:
            C.append(L[i])
            i+=1
        else:
            C.append(R[j])
            j+=1

    while i!=m:
        C.append(L[i])
        i+=1

    while j!=n:
        C.append(R[j])
        j+=1
    
    return C


def ternaryMergeSort(A, low,high):
    if (low  < high): 
        third = low+(high-low)//3
        twoThird = low+2*(high-low)//3
        ternaryMergeSort(A, low, third)
        ternaryMergeSort(A,third+1,twoThird)
        ternaryMergeSort(A, twoThird+1,high)
        A[low:twoThird+1] = merge( A[low:third+1], A[third+1:twoThird+1])
        A[low:high+1] = merge( A[low:twoThird+1], A[twoThird+1:high+1])

    
A1 = []
A2 = [1]
A3 = [1,2]
A4 = [2,1]
A5 = [5,6,1,3,2,1,7,9,5,15,100,2,17,56]
A6 = [20-i for i in range(20)]

ternaryMergeSort(A1,0, len(A1)-1)
ternaryMergeSort(A2,0, len(A2)-1)
ternaryMergeSort(A3,0, len(A3)-1)
ternaryMergeSort(A4,0, len(A4)-1)
ternaryMergeSort(A5,0, len(A5)-1)
ternaryMergeSort(A6,0, len(A6)-1)

print(A1)
print(A2)
print(A3)
print(A4)
print(A5)
print(A6)

```

```python
# Problem 7
class ReemDrawing:
    def __init__(self): 
        self.m = 0  # m represents the number of points, initially zero
        self.L = [] # The list of points initially empty (len(L) is equal to m)
        
    def add_point(self,p): # add a point to the drawing
        assert type(p)==Point, 'Bad input!'
        self.L.append(p)
        self.m+=1
                
    def __str__(self):
        if self.m==0:
            return 'Empty drawing.'
        s = ''
        for i in range(self.m-1):
            s+=str(self.L[i])+'-->'
        s+=str(self.L[m-1])
        return s
    
    def copy_drawing(self, other):
        assert type(other)==ReemDrawing, 'Bad Input!'
        from copy import deepcopy
        self.L = deepcopy(other.L)
        self.m = other.m
        
    def plot_drawing(self):
        import matplotlib.pyplot as plt
        x = [pt.x for pt in self.L]
        y = [pt.y for pt in self.L]
        plt.plot(x,y)
        
    def is_connected(self, other):
        assert type(other)==ReemDrawing, 'Bad Input!'
        
        if self.m==0 or other.m==0: #if either of the drawings is empty
            return False
        
        last_point_in_self = self.L[m-1]
        first_point_in_other = other.L[0]
        
        return (last_point_in_self.x == first_point_in_other.x) and (last_point_in_self.y == first_point_in_other.y)
        
    
d1 = ReemDrawing()
d2 = ReemDrawing()

print('d1:', d1)
print('d2:', d2)

m = int(input('Enter the number of points in the drawing: '))

for i in range(m):
    s = input('Enter the x and y values of Point ' + str(i+1) + ' (separated by a space): ' )
    L = s.split()
    x = float(L[0])
    y = float(L[1])
    p = Point(x,y)
    d1.add_point(p)
    

d2.copy_drawing(d1)
print('d1:', d1)
print('d2:', d2)

if(d1.is_connected(d2) or d2.is_connected(d1)):
    print('The two drawings are connected')
else:
    print('The two drawings are not connected')
    
print('The number of points in d1 is', d1.m)

d1.plot_drawing()

    
```
